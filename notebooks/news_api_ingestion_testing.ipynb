```python
import json #convert Python dictionaries into JSON strings
import time #avoid API rate-limit problems, pause execution (sleep)
import requests #call the NewsAPI endpoint
import datetime #used to work with dates

#using datetime to mention the desired dates to retrieve data
date_from = datetime.date(2025, 12, 5)
date_to = datetime.date.today()

#Convert this dates into this format
str_date_from = date_from.strftime("%Y-%m-%d")
str_date_to = date_to.strftime("%Y-%m-%d")


API_KEY = "b2e1bcf1b85340289fe9aa7aca5ba616"
URL = "https://newsapi.org/v2/"


#endpoint desired to retrieve data (API documentation)
endpoint = "everything"

#what topics I am looking for (these are keywords)
query = '(bitcoin OR BTC OR ethereum OR ETH OR crypto OR cryptocurrency OR blockchain OR web3 OR "digital asset" OR stablecoin OR USDT OR USDC OR Binance OR Coinbase OR Kraken OR regulation OR SEC OR hack OR exploit)'


params = {
    "q": query,  #keyword-based filtering (crypto-related news)
    "from":  str_date_from,
    "to": str_date_to,
    "language": "en",
    "sortBy": "publishedAt",
    "apiKey": API_KEY
}


#sends an HTTP GET request
response = requests.get(URL + endpoint, params=params)


if response.status_code == 200:
    data = response.json()
    articles = data["articles"]
    print(articles)

else:
    print(response.status_code)
    try:
        print(response.json())
        
    except Exception:
        print("No data retrieved")
          
    
```
